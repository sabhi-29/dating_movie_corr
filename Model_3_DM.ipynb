```python
# Extracting the personality of the desired characters from the prominent genre movies for each of the class
import pandas as pd

# Define file paths
ratings_file = 'ratings.dat'
users_file = 'users.dat'
movies_file = 'movies.dat'

# Define column names for each file
ratings_cols = ['UserID', 'MovieID', 'Rating', 'Timestamp']
users_cols = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
movies_cols = ['MovieID', 'Title', 'Genres']

# Read data into Pandas DataFrames with explicit encoding
ratings_df = pd.read_csv(ratings_file, sep='::', engine='python', header=None, names=ratings_cols, encoding='latin-1')
users_df = pd.read_csv(users_file, sep='::', engine='python', header=None, names=users_cols, encoding='latin-1')
movies_df = pd.read_csv(movies_file, sep='::', engine='python', header=None, names=movies_cols, encoding='latin-1')

# Merge the DataFrames
merged_df = pd.merge(pd.merge(ratings_df, users_df, on='UserID'), movies_df, on='MovieID')
merged_df
# Map numeric occupation values to occupation names
occupation_mapping = {
    0: "other or not specified",
    1: "academic/educator",
    2: "artist",
    3: "clerical/admin",
    4: "college/grad student",
    5: "customer service",
    6: "doctor/health care",
    7: "executive/managerial",
    8: "farmer",
    9: "homemaker",
    10: "K-12 student",
    11: "lawyer",
    12: "programmer",
    13: "retired",
    14: "sales/marketing",
    15: "scientist",
    16: "self-employed",
    17: "technician/engineer",
    18: "tradesman/craftsman",
    19: "unemployed",
    20: "writer"
}

# Map numeric age values to age groups
age_mapping = {
    1: "Under 18",
    18: "18-24",
    25: "25-34",
    35: "35-44",
    45: "45-49",
    50: "50-55",
    56: "56+"
}

merged_df['Occupation'] = merged_df['Occupation'].map(occupation_mapping)
merged_df['Age'] = merged_df['Age'].map(age_mapping)

# Display the resulting DataFrame
print(merged_df[['UserID', 'Age', 'Gender', 'Occupation', 'Genres', 'Title', 'Rating']])
```

```python
# Create subsets based on user age, occupation, and gender
subsets = merged_df.groupby(['Age', 'Gender'])

# Define a function to find the prominent genre and popular movies within each subset
def find_prominent_genre_and_popular_movies(subset):
    # Calculate genre counts
    genre_counts = subset['Genres'].str.split('|').explode().value_counts()

    # Find the most prominent genre based on the count
    prominent_genre = genre_counts.idxmax()

    # Find popular movies within the most prominent genre
    popular_movies = subset[subset['Genres'].str.contains(prominent_genre)].groupby('Title').size().sort_values(ascending=False).head(5).index.tolist()

    return {'Prominent_Genre': prominent_genre, 'Popular_Movies': popular_movies}

# Create a dictionary to store the results for each subset
subset_results = {}
for group, subset in subsets:
    subset_results[group] = find_prominent_genre_and_popular_movies(subset)

# Convert the dictionary to a Pandas DataFrame
results_df = pd.DataFrame.from_dict(subset_results, orient='index')

# Reset index for better readability
results_df = results_df.reset_index()

# Display the resulting DataFrame
results_df
```

```python
# Now for each of the common movies we will extract the characters present in that movie and 
# their personality type from our MTBI personality dataset
import math
from itertools import permutations
# First storing the characters, gender and personlity type for all of the common movies
new_subcategory = [x.lower() for x in df_selected['subcategory'].values]
df_selected['subcategory'] = new_subcategory
# df_selected[['I', 'N', 'F', 'P', 'E', 'S', 'T', 'J']]
# print(df_selected[df_selected['subcategory'] == 'american beauty'])
# for rows in df_selected[df_selected['subcategory'] == 'american beauty'].values:
#     print(rows[:12])
#     print()
mtbi = {0: 'I', 1: 'N', 2: 'F', 3: 'P', 4: 'E', 5: 'S', 6: 'T', 7:'J'}
valid_personality_type = list(personality_descriptions.keys()) 
```

```python
def movie_char_pers(movie_title):

    ex_dict = {x[1]:[y for y in x[4:12]] for x in df_selected[df_selected['subcategory'] == movie_title].values}
    # print(ex_dict)
    for keys in ex_dict.keys():
        values = ex_dict[keys]
        temp_val_arr = []
        min_perc = 100
        for i in range(len(values)):
            if math.isnan(values[i]) != True:
                temp_val_arr.append(i)
                if values[i] < min_perc:
                    min_perc = values[i]

        temp_val_arr_pers = ''
        for indexs in temp_val_arr:
            temp_val_arr_pers+= mtbi[indexs]
    #     print(temp_val_arr_pers)
        if len(temp_val_arr_pers) != 4:
            ex_dict[keys] = 'No personality'
        else:
            # Now that we have letters in personality checking the valid permutation
            all_permutations = [''.join(x) for x in list(permutations(temp_val_arr_pers))]
            for i in range(len(all_permutations)):
                if all_permutations[i] in valid_personality_type:
    #                 print(keys, all_permutations[i])
                    ex_dict[keys] = all_permutations[i]

    return ex_dict
    
# Test 
# print(movie_char_pers('star wars'))

# Now we run this function for all movies in a given class and maintain a dictionary with personality type counter
# Before that we need to identify the gender for all the characters manually
```

```python
# Getting all movies of interest
movies_of_interest = []
for movie_list in results_df['Movies of Interest'].values:
    for each_movie in movie_list:
        if each_movie not in movies_of_interest:
            movies_of_interest.append(each_movie)
```

```python
# Making a dictionary of character for each movie of interest and their genders
movie_char_list = []
for movies in movies_of_interest:
    print(f"Current movie: {movies}")
    characters = list(movie_char_pers(movies).keys())
    genders = []
    for i in range(len(characters)):
        char_gender = input(f"Enter gender of {characters[i]}: ")
        genders.append(char_gender)
    char_gen_dict = {}
    for i in range(len(characters)):
        char_gen_dict[characters[i]] = genders[i]
    movie_char_list.append([movies, char_gen_dict])
```

```python
# Now that genders have been assigned to each character we will count the personality type for each class
# and keep store it as a dictionary and add to our results_df dataframe
movie_char_list[10][1]['Padmé Amidala'] = 'F'
movie_char_list[10][1]['K-2SO'] = 'M'
new_column = []
all_movies = []
for movies in movie_char_list:
    all_movies.append(movies[0])

# iterate through each row and each movie of interest for that row, get characters and personality from 
# movie_char_pers and for each character of the movie (key of the output of movie_char_pers) check their 
# gender from movie_char_list[movie_index] dictionary and store the personality type and count as a dictionary
# at then end add an additional column to results_df that has that dictionary 
for rows in results_df.values:
    class_gender = rows[1]
    class_movies_of_interest = rows[4]
    gender_of_interest = 0
    if class_gender == 'M':
        gender_of_interest = 'F'
    else:
        gender_of_interest = 'M'
    
    pers_count_dict = {}
    # Iterating through each movie of interest we will get the characters for each movie and the characters with
    # gender of interest, finally we keep a count of all personalities of the characters
    for movie in class_movies_of_interest:
        movie_index = all_movies.index(movie)
        characters_in_movie = movie_char_pers(movie)
        relevant_char = []
#         print(movie_char_list[movie_index][1])
        for each_char in characters_in_movie.keys():
            if movie_char_list[movie_index][1][each_char] == gender_of_interest:
                relevant_char.append(each_char)
        # Now getting the personality of each character
        for each_char in relevant_char:
            char_personality = characters_in_movie[each_char]
            if char_personality in pers_count_dict.keys():
                pers_count_dict[char_personality] += 1
            else:
                pers_count_dict[char_personality] = 1
    
    # Now we have got the personality count for all movies for the class and relevant characters
    # now we will add this dictionary to the list that keeps track of all personality count for each class
    new_column.append(pers_count_dict)

# new_column
```

```python
for i in range(len(new_column)):
    each_dict = new_column[i]
    # Sort the dictionary items based on values
    sorted_items = sorted(each_dict.items(), key=lambda x: x[1], reverse=True)

    # Create a new dictionary from the sorted items
    sorted_dict = dict(sorted_items)
    new_column[i] = sorted_dict
results_df['Personality Count'] = new_column
results_df['Personality Count'].values
```

```python
# results_df
# Finally we will add a new column to the table consisting of the top 2 most prominent personalities for that class
# But we would do this only for row with index 2,3,5,6,7,8,9,11 since they have highest personality count > 3
# For the rest we will get the most prominent personality only
new_column_again = []
for index in range(len(results_df)):
#     print(index)
    if index in [2,3,5,6,7,8,9,11]:
        pers_dict = results_df.values[index][-1]
        values = list(set(pers_dict.values()))
        two_most_frequent = (sorted(set(values), reverse = True))[:2]
    #     print(two_most_frequent)

        # Now that we have the two most frequent personality counts lets get those personalities
        temp_arr = []
        for each_pers in pers_dict.keys():
            if each_pers == 'No personality':
                continue
            if pers_dict[each_pers] in two_most_frequent:
                temp_arr.append(each_pers)
        # Finally appending it to our new additional column
        new_column_again.append(temp_arr)
        
    else:
        # We just get the most prominent personality and not the top two
        pers_dict = results_df.values[index][-1]
        values = list(set(pers_dict.values()))
        most_frequent = (sorted(set(values), reverse = True))[0]
        temp_arr = []
        for each_pers in pers_dict.keys():
#             print(pers_dict[each_pers], most_frequent)
            if each_pers == 'No personality':
                continue
            if pers_dict[each_pers] == most_frequent:
                temp_arr.append(each_pers)
        # Finally appending it to our new additional column
        new_column_again.append(temp_arr)

# new_column_again
```
